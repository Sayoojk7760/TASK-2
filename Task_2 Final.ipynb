```python
#Identifing ouliers using IQR
```

```python
numeric_df = df.select_dtypes(include=['number'])
```

```python
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1
```

```python
outliers = (numeric_df < (Q1 - 1.5 * IQR)) | (numeric_df > (Q3 + 1.5 * IQR))
```

```python
#FIXING USING BY REMOVING OUTLIERS
```

```python
df_no_outliers = df[~outliers.any(axis=1)]
df_no_outliers
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#correlation heatmap
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
```

```python
corr = df[numeric_cols].corr()
```

```python
plt.figure(figsize=(20, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
#the hetamap shows Show correlations between numeric columns.
 #Strong Positive Correlation**:
  # Logged GDP per capita and Ladder score : A strong positive correlation (~0.79) indicates that as gdp increases ,it reflects an increase in the ladder score .
 #Weaker Correlations**:
  # Generosity and Ladder score : There is a weaker correlation (~0.33) between generosity and ladder scores, suggesting that while generosity might play a role, it is less significant than other factors like GDP and social support.
```

```python
# Categorizing the 'Ladder score' into categories (Low, Medium, High)
bins = [0, 3, 5, 10] 
labels = ['Low', 'Medium', 'High']
df['Ladder score category'] = pd.cut(df['ladder score'], bins=bins, labels=labels, right=False)

# Group by 'Ladder score category' and 'Region', and count occurrences
category_distribution = pd.crosstab(df['regional indicator'], df['Ladder score category'])

# Plot the Stacked Bar Chart
category_distribution.plot(kind='bar', stacked=True, figsize=(12, 8), color=['red', 'yellow', 'green'])

# Title and labels
plt.title('Distribution of Ladder Score Categories by Region')
plt.xlabel('Region')
plt.ylabel('Number of Countries')
plt.xticks(rotation=90)  # Rotate region names for better readability
plt.legend(title='Ladder Score Category')
plt.show()
```

```python
# the stacked bar chart shows the distribution of ladder score categories by region
# the x axis shows region and y axis shows number of countries.
# 35 above countries comes under subsaharan africa.
# mostly the central and easten europe, western europe ,east asia and north america and anz are the regions having only high ladder score countries.
# sub-saharian africa region contains the most number of countries.
# north america and anz region contains the least number of countries.
# south asia region contains the countries with low ladder score.
```

```python
#box plot
df.columns = df.columns.str.strip().str.lower()

print(df['ladder score'].isnull().sum())

plt.figure(figsize=(8, 6))
sns.boxplot(x=df['ladder score'], color='skyblue')

plt.title("Distribution and Outliers of Ladder Score")
plt.xlabel("Ladder Score")
plt.show()
```
